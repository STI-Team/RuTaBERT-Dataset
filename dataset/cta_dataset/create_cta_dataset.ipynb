```python
import pandas as pd
import glob

df_list = []

number_of_chunks = len(glob.glob("../collecting/columns_data/data*.csv"))
for i in range(number_of_chunks):
    df_list.append(
        pd.read_csv(f"../collecting/columns_data/data_{i}.csv", sep="⇧", engine="python", quotechar='"', on_bad_lines="warn")
    )

df_data = pd.concat(df_list, axis=0)
```

```python
df_data_filter = df_data[~df_data["column_data"].isna()]
df_data_filter
```

```python
df_data_filter["composite_key"] = df_data_filter["table_id"] + " / " + df_data_filter["column_id"].astype(str)
df_data_filter
```

```python
threshold = 0.78
df_threshold = pd.read_csv(f"../labelling/labelled/labelled_threshold_{threshold}.csv", sep=";")
```

```python
df_threshold.drop(["column_name", "similarity"], axis=1, inplace=True)
```

```python
df_threshold["composite_key"] = df_threshold["table_id"] + " / " + df_threshold["column_id"].astype(str)
```

```python
df_threshold
```

```python
df_joined = df_threshold.merge(df_data_filter[["composite_key", "column_data"]], how="right", on="composite_key")
df_joined.drop(["composite_key"], axis=1, inplace=True)
```

```python
df_joined
```

```python
labels = set()
with open("../labelling/sem_types.txt", mode="r") as f:
    for line in f:
        labels.add(line.lower().strip())

labels = list(labels)
sem_types = dict((labels[i], i) for i in range(len(labels)))
sem_types
```

```python
df_joined["label_id"] = df_joined["label"].progress_apply(lambda x: sem_types[x])
```

```python
df_joined = df_joined[["table_id", "column_id", "label_id", "label", "column_data"]]
df_joined
```
